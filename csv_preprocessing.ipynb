```python
cols = [
    "country_code", "postal_code", "place_name", "admin_name1", "admin_code1",
    "admin_name2", "admin_code2", "admin_name3", "admin_code3",
    "latitude", "longitude", "accuracy"
]

df_zip = pd.read_csv('allCountries.txt', sep='\t', names=cols, dtype=str)

print(df_zip.head())
```

```python
df = pd.read_csv('airbnb-listings.csv', sep = ';')
df.head(5)
```

```python
subset_col = ['ID','Host Since','Property Type','City','State','Country Code', 'Country','Latitude','Longitude','Zipcode','Geolocation']
df = df[subset_col]
```

```python
df['Country Code'].unique()
```

```python
df['Host Since'] = pd.to_datetime(df['Host Since'])
df['host_since_year'] = df['Host Since'].dt.year
df = df.rename(columns={"Property Type": "property_type", "Zipcode": "postal_code","Country Code":"country_code"})
df
```

```python
df
```

```python
country_list = ['IT','CH', 'FR', 'AT', 'NL', 'GE','ES','DE','DK']
df_filtered = df[df['country_code'].isin(country_list)]
```

```python
df_filtered
```

```python
df_filtered['postal_code'] = df_filtered['postal_code'].astype(str).str.strip()
df_filtered['country_code'] = df_filtered['country_code'].astype(str).str.upper()

df_zip['postal_code'] = df_zip['postal_code'].astype(str).str.strip()
df_zip['country_code'] = df_zip['country_code'].astype(str).str.upper()
```

```python
df_zip = df_zip.rename(columns={"latitude": "zip_latitude", "longitude": "zip_longitude"})
df_zip
```

```python
df_filtered[['postal_code','country_code']]
```

```python
df_merged = df_filtered.merge(df_zip[['country_code','postal_code', 'place_name', 'admin_name1', 'zip_latitude','zip_longitude']], on=['postal_code','country_code'], how='left')

df_merged.rename(columns={
    'place_name': 'mapped_city',
    'admin_name1': 'mapped_region'
}, inplace=True)

df_merged
```

```python
df_merged['Country'].unique()
```

```python
df_merged.groupby(['country_code']).count()
```

```python
df_merged = df_merged.rename(columns={"City": "city_orig", "State": "state_orig","mapped_region":"State","mapped_city":"City"})
df_merged
```

```python
df_merged.to_csv('airbnb-short-listings.csv')
```

```python
df_merged_short = df_merged.groupby(['Country', 'host_since_year', 'property_type']).head(5).reset_index(drop=True)
df_merged_short
```

```python
import json
from collections import Counter

df = df_merged 

clusters = []


city_counts = Counter(df["City"])

for (country, host_since_year, property_type, city), group in df.groupby(["Country", "host_since_year", "property_type", "City"]):
    avg_lat = group["Latitude"].mean()
    avg_lon = group["Longitude"].mean()

    clusters.append({
        "country": country,
        "count": len(group),
        "host_since_year": host_since_year,
        "property_type": property_type,
        "city": city,
        "avg_lat": avg_lat,
        "avg_lon": avg_lon
    })

with open("country_clusters.json", "w") as f:
    json.dump(clusters, f, indent=2)
```
